```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

# Load Data
df = pd.read_csv("./updated.csv")

# Select feature(s) for clustering (you can modify this as needed)
X = df[["Customer_ID"]].dropna()  # Remove NaN values if any

# Create the linkage matrix (hierarchical clustering)
linkage_matrix = sch.linkage(X, method='ward')  # 'ward' minimizes variance

# Plot the Dendrogram
plt.figure(figsize=(8, 4))
sch.dendrogram(linkage_matrix, orientation="top", labels=X.index, distance_sort="descending", show_leaf_counts=True)
plt.xlabel("Data Points")
plt.ylabel("Cluster Distance")
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```
